**DATE AND TIME**

```python
#conveting string of date into datetime dtype.
def years(datess):
    import datetime
    return datetime.datetime.strptime(datess,"%Y-%m-%d") 
```

```python
df1["Date"]=df1["Date"].apply(years)
```

```python
df1.info()
```

```python
#get covid cases between date 30 jan to 4 april of year 2020.
df1[(df1["Date"]>="2020-01-30") & (df1["Date"]<="2020-04-04")]
```

```python
#get total number of cured,deaths,confirmed covid cases of year 2020.
df1[["Cured","Deaths","Confirmed"]][(df1["Date"]>="2020-01-01") & (df1["Date"]<="2020-12-31")].sum()
```

```python
#get total number of cured,deaths,confirmed covid cases of year 2021.
df1[["Cured","Deaths","Confirmed"]][(df1["Date"]>="2021-01-01") & (df1["Date"]<="2021-12-31")].sum()
```

```python
#getting year from date column.
def years1(datess):
    import datetime
    return datetime.datetime.strptime(datess,"%Y-%m-%d").year
```

```python
df1["Year"]=df1["Date"].apply(years1)
```

```python
df1["Year"].info()
```

```python
df1
```

```python
#get total numbers of cured,deaths and confirmed covid cases from 2020 and 2021 year.
df1.groupby("Year").sum()
```

```python
#getting months from date column.
def month(datess):
    import datetime
    return datetime.datetime.strptime(datess,"%Y-%m-%d").month
```

```python
df1["Month"]=df1["Date"].apply(month)
```

```python
df1["Month"].info()
```
